### Distribution of Numeric Features

```python
data.hist(bins=10, figsize=(12, 8))
```

```python
# Creating box plot
plt.figure(figsize=(10, 8))

for i, column in enumerate(data.select_dtypes(include=['int64', 'float64']).columns, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(data[column])
    plt.title(column)

plt.tight_layout()
plt.show()
```

### Role Distribution

```python
# Role distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='Role', data=data)
plt.title('Distribution of Roles')
plt.show()
```

### Correlation Matrix

```python
# Correlation Matrix
plt.figure(figsize=(10, 6))
corelation_matrix = data.drop(columns=['Champion Name', 'Role']).corr()
sns.heatmap(corelation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

### Gold Efficiency by Role

```python
# Gold Efficiency by Role
plt.figure(figsize=(10, 6))
sns.boxplot(x='Role', y='Gold Efficiency', data=data)
plt.title('Gold Efficiency by Role')
plt.show()
```

### Base Mana by Role

```python
# Boxplot of Base Mana by Role
plt.figure(figsize=(10, 6))
sns.boxplot(x='Role', y='Base Mana', data=data)
plt.title('Base mana by Role')
plt.show()
```
